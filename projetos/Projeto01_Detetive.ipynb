Esse projeto tem a finalidade de fixar os conhecimentos de Variáveis, Print, Input
e Condicionais, para isso crie um programa que faça 5 perguntas para uma
pessoa sobre um crime. As perguntas são:

• "Você telefonou para a vítima?"

• " Você esteve no local do crime?"

• " Você mora perto da vítima?"

• " Você devia para a vítima?"

• " Você já trabalhou com a vítima?"

O programa deve, no final, emitir uma classificação sobre a participação da
pessoa no crime.
Se a pessoa responder positivamente a:

• 2 questões ela deve ser classificada como "Suspeita",

• Entre 3 e 4 como "Cúmplice"

• 5 como "Assassino".

• Caso contrário, ele será classificado como "Inocente".


```python
contagem = 0 
resposta = ""

while True:
  resposta = str(input("Voce telefonou para a vítima? \n(S/N): ").upper())
  while resposta != "S" and resposta != "N":
    print("")
    print("Você digitou algo errado.")
    resposta = str(input("Voce telefonou para a vítima? \n(S/N): ").upper())
  
  if resposta == "N":
    print("")
    break

  if resposta == "S":
    contagem+= 1
    print("")
    break

while True:
  resposta = str(input("Voce esteve no local do crime? \n(S/N): ").upper())
  while resposta != "S" and resposta != "N":
    print("")
    print("Você digitou algo errado.")
    resposta = str(input("Voce esteve no local do crime? \n(S/N): ").upper())
  
  if resposta == "N":
    print("")
    break

  if resposta == "S":
    contagem+= 1
    print("")
    break

while True:
  resposta = str(input("Você mora perto da vítima? \n(S/N): ").upper())
  while resposta != "S" and resposta != "N":
    print("")
    print("Você digitou algo errado.")
    resposta = str(input("Você mora perto da vítima? \n(S/N): ").upper())
  
  if resposta == "N":
    print("")
    break

  if resposta == "S":
    contagem+= 1
    print("")
    break

while True:
  resposta = str(input("Você devia para a vítima? \n(S/N): ").upper())
  while resposta != "S" and resposta != "N":
    print("")
    print("Você digitou algo errado.")
    resposta = str(input("Você devia para a vítima? \n(S/N): ").upper())
  
  if resposta == "N":
    print('')
    break

  if resposta == "S":
    contagem+= 1
    print("")
    break

while True:
  resposta = str(input("Você já trabalhou com a vítima? \n(S/N): ").upper())
  while resposta != "S" and resposta != "N":
    print("")
    print("Você digitou algo errado.")
    resposta = str(input("Você já trabalhou com a vítima? \n(S/N): ").upper())
  
  if resposta == "N":
    print("")
    break

  if resposta == "S":
    contagem+= 1
    print("")
    break

if contagem == 5:
  print(f"Você é o assassino! {contagem} pontos.")
  
if contagem == 3 or contagem == 4:
  print(f"Você é cúmplice! {contagem} pontos.")
    
if contagem == 2:
  print(f"Você é suspeito! {contagem} pontos.")

if contagem == 1 or contagem < 1:
  print(f"Você é inocente! {contagem} pontos.")
```
